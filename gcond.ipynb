**ANS: 生成的凝結圖（Condensed Graph）剩下 70 個節點。**   

size of adj_train: (140, 140)          
edges in adj_train: 42.0            
adj_syn: (70, 70) feat_syn: torch. Size([70, 1433])   


推論原因：

1.   從 Log 觀察： 程式執行結果顯示 adj_syn: (70, 70)。在圖學中，鄰接矩陣（Adjacency Matrix）的維度為 $(N \times N)$，其中 $N$ 代表節點數量。因此 $(70, 70)$ 的矩陣意味著圖中有 70 個節點。
2.   從參數驗證： 本次實驗設定的壓縮率（Reduction Rate, r）為 0.5（指令中的 --r=0.5）。原始訓練圖有 140 個節點，計算 $140 \times 0.5 = 70$，與生成的矩陣大小相符。

上面的範例是論文作者使用一種GCN模型來訓練CORA資料集，然後透過gradient matching的方法(這裡用的是GCond)製作一個合成圖，之後我使用一個自製的GCN模型對合成圖進行測試但發現準確度和論文的方法相比奇差無比(剩下30%左右)，請簡述一下可能的原因。

(Hint:使用GCond方法合成圖時，是不是會有模型綁定的問題?)



**ANS: 模型耦合 (Model Coupling/Binding) 或 對特定架構的過擬合。**  


GCond 的核心機制是 Gradient Matching，優化目標是讓「合成圖在特定模型產生的梯度」去逼近「原始圖在該模型產生的梯度」，因此生成的合成圖 $S$ 實際是為了欺騙那個特定的 GCN 架構而生成的，它將資訊編碼成了該特定模型容易讀取的特徵分佈。

當您換了一個自製的 GCN 模型，在層數、激活函數或聚合方式都可能不同的情況下，模型在參數空間中的梯度方向會完全不同，原本合成圖中隱藏的特徵資訊對新模型來說可能變成了雜訊或無法解讀的數據，導致準確率大幅下降。
